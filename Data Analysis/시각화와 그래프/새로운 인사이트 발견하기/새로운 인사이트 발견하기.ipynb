## 실습 : 직업 탐구하기 2
groupby 문법을 사용해서 '여성 비율'이 높은 순으로 직업을 나열하기,
Dataframe이 아닌 Series로, 'gender'에 대한 값만 출력

```python
oc_df['gender'] = pd.to_numeric(oc_df['gender']) # oc_df['gender']의 타입이 object로 반환되는데 이것을 int타입으로 바꿔줘야됨
```

```python
# 'gender'가 문자로 되어있기때문에 평균을 구할 수 없기 때문에 'M'=0, 'F'=1로 변경하여 숫자로 계산이 가능하게 함
oc_df.loc[oc_df['gender'] == 'M', 'gender'] = 0
oc_df.loc[oc_df['gender'] == 'F', 'gender'] = 1
```

```python
oc_gb = oc_df.groupby('occupation')
```

```python
oc_gb.mean()['gender'].sort_values(ascending=False)
```

이제 각 직업별로 성별 비율을 알 수 있게 되었습니다.
남자가 숫자 0, 여자가 숫자 1이기 때문에, 평균이 0에 가까울 수록 남자가 많고, 1에 가까울 수록 여자가 많다는 뜻입니다.

즉, 평균값이 0.250000이라면 25퍼센트가 여성이고, 75퍼센트가 남성입니다.

결과를 분석해보면,

homemaker의 경우 여성 비율이 86퍼센트 정도 되고, healthcare는 여성 비율이 69퍼센트 정도 되고, 
반대로, doctor는 100퍼센트가 남성이며, engineer는 3퍼센트 정도만 제외하고 모두 남성임을 알 수 있습니다.
