# 데이터 클리닝 : 완결성

```python
import pandas as pd
```

```python
at_df = pd.read_csv('attendance.csv', index_col=0)
at_df
```

```python
at_df.isnull() # 결측값은 True로 나옴
```

```python
at_df.isnull().sum() 
```

```python
# 결측값 지우기
# 기존 데이터 프레임을 건드리지않고 새로운 데이터 프레임을 만듬
# inplace=True로 기존에 반영할 수 있음
at_df.dropna()
```

```python
# 결측값이 있는 컬럼 지우기
at_df.dropna(axis='columns')
```

```python
# 결측값 부분을 다른 값으로 대체
at_df.fillna(0)
```

```python
# 평균값이나 중간값으로 대체
at_df.fillna(at_df.mean())
```

# 데이터 클리닝 : 유일성

```python
dust_df = pd.read_csv('dust.csv', index_col=0)
dust_df.head(10)
```

```python
# row 중복일때
dust_df.index.value_counts()
```

```python
dust_df.loc['07월 31일']
```

```python
# 중복된 row 삭제
dust_df.drop_duplicates(inplace=True)
```

```python
dust_df
```

```python
# 중복된 column 삭제 
## T  -> row와 column 바꿈
dust_df = dust_df.T.drop_duplicates().T
```
