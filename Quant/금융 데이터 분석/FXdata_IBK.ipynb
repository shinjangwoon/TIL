# 기업은행 환율 조회
https://www.ibk.co.kr/fxtr/excRateList.ibk

- 화면상 10초 이내 다시 요청 방지 (크롤링에는 영향 없음)
- pd.read_html(url)로 간단하게 조회 가능

```python
import pandas as pd

url = 'https://www.ibk.co.kr/fxtr/excRateList.ibk?inDate=2021.06.21' # 영업일만 가능
dfs = pd.read_html(url)
df = dfs[0]
df
```

# 컴마(,) 삭제

```python
df['현찰살때(스프레드)'] = df['현찰살때(스프레드)'].str.replace(',', '')
df['현찰팔때(스프레드)'] = df['현찰팔때(스프레드)'].str.replace(',', '')
```

# 데이터 분할
- 현찰살때(스프레드) → 현찰살때, 현찰살때_스프레트
- 현찰팔때(스프레드) → 현찰팔때, 현찰팔때_스프레트

```python
df['현찰살때'] = df['현찰살때(스프레드)'].str.extract(r'([\.\d]+)', expand=False)
df['현찰살때_스프레드'] = df['현찰살때(스프레드)'].str.extract('\((.*?)%\)', expand=False)

df['현찰팔때'] = df['현찰팔때(스프레드)'].str.extract(r'([.\d]+)', expand=False)
df['현찰팔때_스프레드'] = df['현찰팔때(스프레드)'].str.extract('\((.*?)%\)', expand=False)
```

# 컬럼 삭제
- 현찰살때(스프레드), 현찰팔때(스프레드) 컬럼 삭제

```python
df = df.drop(['현찰살때(스프레드)', '현찰팔때(스프레드)'], axis=1)
df
```

# 함수로 정리

```python
import pandas as pd
from datetime import datetime

def FX_IBK(date=None):
    date = datetime.today() if date == None else date
    date = date.strftime('%Y.%m.%d') if type(date) == datetime else date 
    
    url = 'https://www.ibk.co.kr/fxtr/excRateList.ibk?inDate=' + date
    dfs = pd.read_html(url)
    df = dfs[0]

    df['현찰살때(스프레드)'] = df['현찰살때(스프레드)'].str.replace(',', '')
    df['현찰팔때(스프레드)'] = df['현찰팔때(스프레드)'].str.replace(',', '')

    df['현찰살때'] = df['현찰살때(스프레드)'].str.extract(r'([\.\d]+)', expand=False)
    df['현찰살때_스프레드'] = df['현찰살때(스프레드)'].str.extract('\((.*?)%\)', expand=False)

    df['현찰팔때'] = df['현찰팔때(스프레드)'].str.extract(r'([.\d]+)', expand=False)
    df['현찰팔때_스프레드'] = df['현찰팔때(스프레드)'].str.extract('\((.*?)%\)', expand=False)

    df = df.drop(['현찰살때(스프레드)', '현찰팔때(스프레드)'], axis=1)

    numeric_cols = ['매매기준율', '송금보낼때', '송금받을때', 
                    '현찰살때', '현찰살때_스프레드', '현찰팔때', 
                    '현찰팔때_스프레드', 'T/C살때', '미화환산율']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df = df.set_index('통화')
    return df
```

```python
df = FX_IBK(datetime(2021, 6, 21))
df
```
