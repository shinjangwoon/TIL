# 반복문
반복문은 자료구조의 데이터 수만큼 코드를 반복실행 합니다.  
자료구조에 있는 "데이터를 하나씩 바인딩하고 들여쓰기된 코드를 실행" 과정을 반복하는 겁니다. 
  
## 반복문 - 기본형
`for 변수 in 자료구조 : ` 형태로 코드를 작성합니다. 코드 끝에 콜론 `:` 있음에 주의하세요. 분기문과 같이 콜론 다음에는 들여쓰기와 함께 영향을 받는 코드를 작성합니다.

###### 반복문을 사용하지 않고 한 줄에 하나의 값 출력하기

````python
리스트 = ["가", "나", "다", "라", "a", "b", "c"]
print(리스트[0])
print(리스트[1])
print(리스트[2])
print(리스트[3])
````

###### 반복문으로 코드를 정리하기

````python
for 변수 in 리스트:
    print(변수)
````

````python
리스트 = ["철수", "영희", "희철"]

for 승객 in 리스트:
    print(f"{승객} 바보")
````

- 기존에 정의한 `n` 변수에 정의된 값이 출력된다. 

````python
n = 10
````

````python
list = [1,2,3,4,5]
for i in list:
    print(n)
````

다음의 반복문은 아래의 코드를 축약해서 쓴것과 같다

````python
for i in [1, 2, 3]:
    print(i)
````

````python
i = 1
print(i)

i = 2
print(i)

i = 3
print(i)
````

파이썬에서는 데이터 타입이 중요하다. `num + 1`을 실행하면 `None`과 숫자를 더하지 못해 에러가 발생한다. 

````python
list = [1,2,3]
i = 10
for i in list:
  if i != 3:
    print(i)

print(i)
````

````python
i = 10
i = 1    # 리스트 0번 데이터
if i != 3:
    print(i)

i = 2    # 리스트 1번 데이터
if i != 3:
    print(i)

i = 3    # 리스트 2번 데이터
if i != 3:
    print(i)
````
