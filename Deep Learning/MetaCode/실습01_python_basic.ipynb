# 함수
- Basic

```python
def calc(x, y, op='+'):
    if op == '+':
        z = x + y
    elif op == '-':
        z = x - y
    elif op == '*':
        z = x * y
    elif op == '/':
        z = x / y
    else:
        z = 'This operation is not defined'
    
    return z
```

```python
print (calc(5,5,'/'))
```

```python
def return_add_minus(x,y):
    return x+y, x-y
```

```python
plus, minus = return_add_minus(5,3)
print (plus)
print (minus)
```

```python
class Cls:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def add(self):
        print (self.x + self.y)
        
    def sub(self):
        print (self.x - self.y)
        
    def mul(self):
        print (self.x * self.y)
        
    def div(self):
        print (self.x / self.y)
```

```python
f = Cls(5,3)
f.add()
f.sub()
f.mul()
f.div()
```

```python
class Cls1:
    def __init__(self, cls1_val):
        self.cls1_val = cls1_val
        
    def view(self):
        print (self.cls1_val)
        
class Cls2(Cls1):
    def __init__(self, cls1_val, cls2_val):
        super(Cls2, self).__init__(cls1_val)
        self.cls2_val = cls2_val
    
    def view(self):
        print(self.cls1_val, self.cls2_val)
```

```python
f1 = Cls1(3)
f1.view()
f2 = Cls2(3, 5)
f2.view()
```
